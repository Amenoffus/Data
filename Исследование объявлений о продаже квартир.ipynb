В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Предобработка данных

### Пропуски в данных

```python
data.isnull().sum().sort_values()
```

Пропуски в процентном соотношении

```python
data.isnull().sum().sort_values()/len(data)*100
```

Удаление пропусков в столбцах 'locality_name' и 'floors_total'.

Количество пропущенных значений незначительно. Пропуски могут быть связаны с проблемами выгрузки или могли не указать значения. Заменить названия локаций не получится, а количество этажей можно узнать по объявлениям из того же дома, но так как данных нет то меньше проблем будет если удалить значения.

```python
data = data.dropna(subset=['locality_name', 'floors_total'])
```

Занчения в столбцах 'is_apartment' и 'balcony' луше заполнить на 'False' и '0' соответственно. 

Количество пропусков в этих столбцах достаточно большое и проигнорировать их нельзя. Возможно, данные не были изначально заполнены или могли возникнуть проблемы с выгрузкой. Пропуски в столбце с балконами могут быть связаны с отсутствием балконов у недвижимости.



```python
data['is_apartment'].unique()
```

```python
data['is_apartment']=data['is_apartment'].fillna(False)
```

```python
data['balcony'] = data['balcony'].fillna(0)
```

Пропуски в столбцах 'living_area' и 'kitchen_area' можно заполнить, посчитав соотношения жилой площади к общей и площади кухни к общей, взять медианы этих значений и применить их для расчета пропущенных значений.

```python
print('Количество пропущенных значений в столбце "living_area":', data.living_area.isnull().sum())
print('Количество пропущенных значений в столбце "kitchen_area":', data.kitchen_area.isnull().sum())
```

```python
lm = data.living_area / data.total_area
lm = round(lm.median(), 2)
lm
```

```python
km = data.kitchen_area / data.total_area
km = round(km.median(),2)
km
```

```python
data.living_area = data.living_area.fillna(data.total_area * lm)
data.kitchen_area = data.kitchen_area.fillna(data.total_area * km)
```

```python
print('Количество пропущенных значений в столбце "living_area":', data.living_area.isnull().sum())
print('Количество пропущенных значений в столбце "kitchen_area":', data.kitchen_area.isnull().sum())
```

Пропуски в столбцах:

- 'ceiling_height'

   - Достаточно большое количество пропусков в этом столбце. Значения могли быть не указаны изначально или есть проблема с выгрузкой данных. Заполнить значений можно было бы из других объявлений в тех же домах, но поскольку таких данных нет лучше оставить их без изменений так, как они могут исказить дальнейший анализ.  

- 'ponds_around3000', 'parks_around3000', 'cityCenters_nearest', 'airports_nearest', 'ponds_nearest', 'parks_nearest', 'days_exposition'
    - Достаточно большое количество пропусков может быть связано с невозможностью расчиать растояние до ближайшего центра, парка или аэропорта, а так же с проблемами в выгрузке данных. Отсутствие дней показа может говорить о том, что недвижимость все еще продается.

## Расчёты и добавление результатов в таблицу

```python
# Создание столбца с ценой за квадратный метр

data['square_meter_price'] = data['last_price'] / data['total_area']
```

```python
# Создание столбцов с днем недели, месяцем и годом публикации объявления

data['day_of_the_week'] = pd.DatetimeIndex(data['first_day_exposition']).dayofweek
data['month'] = pd.DatetimeIndex(data['first_day_exposition']).month
data['year'] = pd.DatetimeIndex(data['first_day_exposition']).year
```

```python
# Создание столбца с категорией квартиры по этажам

def floor_category(row):
    if row['floor'] == 1: 
        return 'Первый'
    if row['floor'] == row['floors_total']:
        return 'Последний'
    else:
        return 'Другой'
    
data['floor_category'] = data.apply(floor_category, axis = 1)
```

```python
# Создание столбцов с соотношением жилой и кухонной площадей к общей площади квартиры

data['living_area_ratio'] = data['living_area'] / data['total_area']
data['kitchen_area_ratio'] = data['kitchen_area'] / data['total_area']
```

```python
def hist_sr(column, bins):
    ax = data_center.plot(kind = 'hist',
                          y = column,
                          histtype = 'step',       # данный тип удобен для сравнения двух гистограмм
                          bins = bins,
                          alpha = 0.5,
                          linewidth = 2,           # ширина графика
                          grid = True,
                          density = True,          # отобразить плотность распределения данных
                          legend = True,           # отобразить пояснение для графиков
                          label = column + ' center_spb'  # строка подписи для графиков
                         )
    data.plot(kind = 'hist',
                            y = column,
                            histtype = 'step',
                            ax = ax,               # отобразить график в осях предыдущей гистограммы
                            bins = bins,
                            alpha = 0.5,
                            linewidth = 2,
                            grid = True,
                            density = True,
                            legend = True,
                            label = column + ' all'
                           )
    plt.show()
```
